```python
from xml.etree import ElementTree as ET

tree = ET.parse('plant_catalog.xml')

root = tree.getroot()

for child in root:
    print(child.tag)
    print(child.text)
    for grandchild in child:
        #print(grandchild.tag)
        print(grandchild.text)
```

```python
root[1].remove(root[1][0])
```

```python
# adding grandchildren
common = ET.Element('Common')
root[1].append(common)
common.text = '2' # note- only applies this to the last <common>
```

```python
for child in root:
    print(child.tag)
    print(child.text)
    for grandchild in child:
        #print(grandchild.tag)
        print(grandchild.text)
```

```python
# Print out all the values given to the grandchildren of tag <common>
for child in root:
    common = child.find('COMMON')

    if common is not None:
        print(common.text)
    else:
        print("no common found")
```

```python
# finds the amount of 'PRICE' tags
price_elements = len(root.findall('.//PRICE'))
print(price_elements)
```
