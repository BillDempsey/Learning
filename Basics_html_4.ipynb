```python
from bs4 import BeautifulSoup
import requests

url = "https://coinmarketcap.com/"
result = requests.get(url).text
doc = BeautifulSoup(result, "html.parser")
```

```python
tablebody = doc.tbody # found it by inspecting the tree
table_rows = tablebody.contents
print(table_rows)
```

```python
print(table_rows[0].next_sibling)
print(table_rows[1].previous_sibling)
```

```python
print(table_rows[0].parent.name)
```

```python
# all descendants
print(table_rows[0].descendants)
```

```python
# just tag names which are children of table_rows[0]
print(list(table_rows[0].children))
```

```python
prices = {}
for row in table_rows[:10]:
    for data_point in row.contents[2:4]:
        print(data_point)
        print("")
```

```python
for row in table_rows[:10]:
    name, price = row.contents[2:4]
    # print(name)
    # print("")

    # have now figured out the name (i.e. Bitcoin) is inside the <p> tag, so...
    #print(name.p.string)
    fixed_name = name.p.string
    # same for prices, just inside the <span> tag
    fixed_price = price.span.string

    # store this all in a dictionary
    prices[fixed_name] = fixed_price

print(prices)
```
